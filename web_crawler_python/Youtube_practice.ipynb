```python
import requests
from bs4 import BeautifulSoup
import json
import re
import urlparse
import shutil
```

```python
res = requests.get("https://www.youtube.com/watch?v=yaL9aecmv_U")
soup =  BeautifulSoup(res.text,"lxml")
name = soup.select("title")[0].text
print name[1:5]
```

```python
m = re.search('"args":({.*?})',res.text)
print m.group(1) #匹配在()中的子字符串
```

```python
jd = json.loads(m.group(1)) 
print jd["url_encoded_fmt_stream_map"]
```

```python
a = urlparse.parse_qs(jd["url_encoded_fmt_stream_map"])
print a
```

```python
print a['url']
```

```python
print a['url'][0]
```

```python
res2 = requests.get( a['url'][0],stream = True)
names = name[0:5] + ".mp4"
f = open(names,"wb")
shutil.copyfileobj(res2.raw,f)
f.close()
```
